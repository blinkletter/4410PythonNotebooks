```python
import numpy as np
import pandas as pd

import uncertainties as un
import uncertainties.unumpy as unp


df = pd.read_csv('TaftData1952.csv',
                 delimiter = ",", 
                 skipinitialspace=True, 
                 index_col="Sub", 
                 comment = "#") 

```

```python
n = df['base ratio']
e = df['base error']

p = unp.uarray(n, e)  # Create an array of uncertainties

#a = 10**p * 43E-6  # Convert to an array of values with uncertainties
a = 10**p * 0.112  # Convert to an array of values with uncertainties

unp.log10(a/a[1])
```

```python
for i in a:
#    print(f"{i.n:.3e} Â± {unp.std_devs(i):.3e}")
    print(f"{i:.3e}")
```
