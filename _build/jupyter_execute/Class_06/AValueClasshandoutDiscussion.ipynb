## Problem 1

The equilibrium constant for the axial conformer of methyl cyclohexene over the equatorial conformer is about 0.5. Calculate the difference in free energy between the two conformations.

$$K_{eq} = e^{-\frac{\Delta G^{\circ}}{RT}}$$

Therefore

$$\Delta G^{\circ} = RT \ln{K_{eq}}$$

```python
### Calculate Free energy from Keq

Keq = 0.5

R = 8.314  # J/mole K
T = 298    # K

StandardDeltaG = R * T * np.log(Keq) / 1000

print(f"The diffence in standard free energy is {StandardDeltaG:+.2f} kJ/mole")

print(f"The A-value for the Me group is {df1['Avalue']['Me']} kJ/mole")
```
